```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import pandas as pd
nltk.download('punkt')
nltk.download('popular')
```

```python
data = pd.read_csv('NPS_research_summaries.csv')
```

```python
summaries = data['Summary']
```

```python
summaries.to_csv("summaries.csv")
```

```python
#this block of code will tokenize each summary and remove stopwords
sum_series = []

for summary in summaries:
    if type(summary) == type(""): #if summary is a string
        summary = summary.lower()
    else:
        continue
    tokenizer = RegexpTokenizer(r'\w+') #use regex to only select words (remove punctuation)
    tokens = tokenizer.tokenize(summary) #get a list of the individual words in the summary
    
    stop_words = set(stopwords.words('english'))
    filtered_summary = []

    #remove stop words
    for w in tokens:
        if w.lower() not in stop_words:
            filtered_summary.append(w)

    #print("filtered_summary")
    #print(filtered_summary)

    s = pd.Series(filtered_summary)
    sum_series.append(s.value_counts()) #get the counts of each unique word and store it to be aggregated later
    #print(s.value_counts())
    #print("\n")
```

```python
#this loop will generate the word count matrix from the words counts in sum_series
df1 = pd.DataFrame()
i = 0
for s in sum_series:
    df2 = s.to_frame(name="summary" + str(i))
    df1 = pd.concat([df1, df2], axis=1)
    i += 1
```

```python
df1["sum"] = df1.sum(axis=1) #add a sum column
```

```python
df1 #final word count matrix dataframe
```

```python
sort = df1["sum"].sort_values(ascending=False)
sort.to_csv("word_counts.csv")
```

```python
df1_t = df1.transpose()
df1_t.to_csv("word_count_matrix_all.csv")
```
